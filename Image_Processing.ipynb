```python
import cv2
```

```python
imgCV = cv2.imread("lena.png",cv2.IMREAD_GRAYSCALE)
```

```python
imgIO = io.imread("lena.png",as_gray=True)
```

```python
img = imgCV
```

Apply Gaussian , Sobel , Median

```python
from skimage.filters import gaussian,sobel,median
```

```python
img_gauss = gaussian(img,sigma=1,mode="constant",cval=0.02)
```

```python
io.imshow(img_gauss)
```

```python
img_sobel = sobel(img)
io.imshow(img_sobel)
```

```python
img_median = median(img)
io.imshow(img_median)
```

Edge Detection :

```python
edge_img = cv2.imread("Osteosarcoma_01.jpeg",cv2.IMREAD_GRAYSCALE)
edge = cv2.Canny(edge_img,120,130)
io.imshow(edge)
```

Merge Image

```python
img1 = cv2.imread("Flower1.jpg")
img2 = cv2.imread("Flower2.jpeg")
```

```python
img1.shape
```

```python
img2.shape
```

```python
img1 = cv2.resize(img1,(img2.shape[1],img2.shape[0]))
```

```python
img1.shape
```

```python
img_merged = cv2.addWeighted(img1,0.5,img2,0.5,0)
io.imshow(img_merged)
```
